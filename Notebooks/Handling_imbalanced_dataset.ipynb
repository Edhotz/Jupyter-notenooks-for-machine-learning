This is a mighty imbalanced dataset


0 -> Legit Transactions
1 -> Fraudlent Transactions


```python
# Separating the legit and fraudlent transactions

legit = credit_card_df[credit_card_df.Class == 0]
fraud = credit_card_df[credit_card_df.Class == 1]
```

```python
print(legit.shape)
```

```python
print(fraud.shape)
```

Under-sampling


### Build a sample dataset simular distribuition of legit and fraudlent transactions

##### Number of #Fraudlent transactions -> 492


```python
legit_sample = legit.sample(n = 492)
legit_sample
```

```python
print(legit_sample.shape)
```

#### Concatenate the two dataframes


```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
new_dataset
```
